# ¿Qué es programación funcional?

La programación funcional busca expresar transformaciones de datos de forma compacta y declarativa.

## Imperativo
Se enfoca en:
- cómo hacerlo

## Funcional
Se enfoca en:
- qué transformar


````python

reviews = [
    " Excelente producto ",
    "MALO",
    "",
    "  recomendado "
]

print(reviews)
````


# Ejemplo: enfoque imperativo


````python

clean = []

for r in reviews:
    if r.strip() != "":
        clean.append(r.strip().lower())

print(clean)
````


# Ejemplo: enfoque funcional


````python

clean = [
    r.strip().lower()
    for r in reviews
    if r.strip() != ""
]

print(clean)
````


# Comprensión de listas (List Comprehension)

La comprensión de listas es una forma compacta de transformar datos.


````python

numbers = [1,2,3,4,5]

squares = [x**2 for x in numbers]

print(squares)
````


# Ejercicio 1

Convertir todos los nombres a formato lowercase (todos en minúsculas) y sin espacios redundantes:

```python
['Juan', 'Ana', 'Pedro', 'Luis']
```


````python

names = [
    "SELENA"
    " KAROL",
    "andres ",
    " nATALIA ",
    "Robert Miles"
]

# Resultado esperado:
# ["selena", "karol", "andres", "natalia", "robert miles"]

# Tu solución aquí

resultado = list(map(lambda x: x.strip().lower(), names))
print(resultado)

````


# Ejercicio 1.1

Convertir todos los nombres a formato Sentence case (primera en mayúsculas) y sin espacios redundantes::

````python

names = [
    "SELENA",
    " KAROL",
    "andres ",
    " nATALIA ",
    "Robert Miles"
]

# Resultado esperado:
# ["Selena", "Karol", "Andres", "Natalia", "Robert miles"]

# Tu solución aquí

resultado = list(map(lambda x: x.strip().title(), names))
print(resultado)
````


# map()

`map()` permite transformar elementos de una colección.


````python

names = [" SELENA ", "karol", "NaTaLiA"]

clean = list(
    map(lambda x: x.strip().title(), names)
)

print(clean)
````


# Ejercicio 2

Calcular la longitud de cada comentario.


````python

reviews = [
    "excelente producto",
    "muy malo",
    "normal"
]

# Resultado esperado:
# [18, 8, 6]

# Tu solución aquí

resultado = list(map(lambda x: len(x), reviews))
print(resultado)
````


# Ejercicio 2.1

Crear un diccionario en el cual la llave sea el comentario y el valor sea la longitud del comentario

````python

reviews = [
    "excelente producto",
    "muy malo",
    "normal"
]

# Resultado esperado:
# {"excelente producto": 18, "muy malo": 8, "normal": 6]

# Tu solución aquí

resultado = dict(map(lambda x: (x, len(x)), reviews))
print(resultado)
````


# filter()

`filter()` permite filtrar elementos según una condición.


````python

reviews = [
    "excelente",
    "",
    "muy malo",
    " ",
    "recomendado"
]

valid = list(
    filter(lambda x: x.strip() != "", reviews)
)

print(valid)
````


# Ejercicio 3

Filtrar comentarios con más de 10 caracteres.


````python

comments = [
    "excelente producto",
    "ok",
    "muy recomendable",
    "malo"
]

# Tu solución aquí

resultado = list(filter(lambda x: len(x) > 10, comments))
print(resultado)
````


# Expresiones lambda

Las funciones lambda permiten crear funciones pequeñas y rápidas.


````python

square = lambda x: x**2

print(square(5))
````


# Ejercicio 4

Clasificar ratings:

- rating >= 4 → positivo
- rating <= 2 → negativo
- otro → neutral


````python

ratings = [5,4,2,1,3]

# Resultado esperado:
# ['positivo','positivo','negativo','negativo','neutral']

# Tu solución aquí

resultado = list(map(lambda x: "positivo" if x >= 4 else ("negativo" if x <= 2 else "neutral"), ratings))
print(resultado)
````


# Tokenización básica

Separar texto en palabras usando `split()`.


````python

text = "excelente producto muy útil"

tokens = text.split()

print(tokens)
````


# Ejercicio 5

Tokenizar todos los comentarios.


````python

reviews = [
    "excelente producto",
    "muy lento",
    "me encantó"
]

# Resultado esperado:
# [
#   ['excelente','producto'],
#   ['muy','lento'],
#   ['me','encantó']
# ]

# Tu solución aquí

resultado = list(map(lambda x: x.split(), reviews))
print(resultado)
````


# Conteo de palabras

Esto nos conectará para el siguiente módulo del curso: NLP.


````python

words = [
    "excelente",
    "malo",
    "excelente",
    "rápido"
]

frequency = {}

for word in words:
    frequency[word] = frequency.get(word, 0) + 1

print(frequency)
````


# Ejercicio 6

Contar frecuencia de palabras usando comprensión de diccionarios o lógica funcional.


````python

words = [
    "excelente",
    "malo",
    "excelente",
    "rápido",
    "malo",
    "excelente"
]

# Tu solución aquí

frecuencia = {palabra: words.count(palabra) for palabra in words}
print(frecuencia)
````


# reduce()

`reduce()` permite acumular valores.

⚠️ DATO IMPORTANTE!!! En ciencia de datos moderna muchas veces Pandas reemplaza este tipo de operaciones.


````python

from functools import reduce

nums = [1,2,3,4]

total = reduce(lambda a,b: a+b, nums)

print(total)
````


# Ejercicio Final — Pipeline funcional

Objetivos:
1. eliminar vacíos
2. quitar espacios
3. convertir a minúsculas
4. tokenizar
5. calcular cantidad de palabras


````python

reviews = [
    " Excelente producto ",
    "",
    "MALO",
    "Muy rápido",
    "  recomendado "
]

# Salida esperada:
# [
#   {'review':'excelente producto', 'words':2},
#   {'review':'malo', 'words':1},
#   {'review':'muy rápido', 'words':2},
#   {'review':'recomendado', 'words':1}
# ]

# Tu solución aquí

reviews = [
    " Excelente producto ",
    "",
    "MALO",
    "Muy rápido",
    " recomendado "
]

resultado = list(
    map(
        lambda r: {
            'review': r.strip().lower(),
            'words': len(r.strip().split())
        },
        filter(lambda r: r.strip() != "", reviews)
    )
)

print(resultado)
````
