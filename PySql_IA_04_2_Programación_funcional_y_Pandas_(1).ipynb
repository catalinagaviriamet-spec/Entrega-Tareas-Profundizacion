# **1.1  Descargar archivo con los datos y convertirlo a DataFrame**

# **2.5  Días críticos**

La columna "new_deaths" indica el número de muertes para la fecha específica.

Incluir una columna llamada "critical_day" si la cantidad de muertes de ese día fue mayor que 500.

Imprimir los primeros registros del DataFrame, solo con las columnas "location", "population", "new_deaths" y "critical_day", para corroborar que el cálculo de la nueva columna "critical_day" haya quedado bien

```python
#Escribir aquí el código para mostrar lo solicitado

df['critical_day'] = df['new_deaths'].apply(lambda x: 'YES' if x > 500 else 'NO')

df.loc[
    :,
    ['country', 'population', 'new_deaths', 'critical_day']
].head()
```
